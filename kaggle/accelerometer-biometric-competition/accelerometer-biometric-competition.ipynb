```python
import numpy as np
label_val = pd.unique(df_train['Device'])
label_map = {}
for i, val in enumerate(label_val):
    label_map[val] = i
label_map
```

```python
df_train['label'] = df_train['Device'].map(label_map)
df_train['space'] = np.sqrt(df_train['X']*df_train['X']+df_train['Y']*df_train['Y']+df_train['Z']*df_train['Z'])

df_train
```

```python
df_train_group = df_train.groupby('label')
fea_col = ['X', 'Y', 'Z', 'space']
X_train, Y_train = [], []
for key, df_group in df_train_group:
    group_val = df_group[fea_col].values
    for i in range(0, len(df_group)-300, 150):
        tmp = group_val[i:i+300,]
        if tmp.shape[0] != 300: print(tmp.shape)
        X_train.append(tmp)
        Y_train.append(key)
# X_train, Y_train = np.array(X_train), np.array(Y_train)
# X_train.shape, Y_train.shape
```

```python
X_train = np.array(X_train)
X_train.shape
```

```python
from torch.utils.data import DataLoader,TensorDataset
import torch
X_train = torch.tensor(X_train, dtype=torch.float)
Y_train = torch.tensor(Y_train, dtype=torch.float)
trainloader = DataLoader(TensorDataset(X_train, Y_train), 128, True)
```

```python
import numpy as np
df_test['label'] = df_test['Device'].map(label_map)
df_test['space'] = np.sqrt(df_test['X']*df_test['X']+df_test['Y']*df_test['Y']+df_test['Z']*df_test['Z'])

df_test
```

```python
# df_test_group = df_test.groupby('label')
fea_col = ['X', 'Y', 'Z', 'space']
X_test, Y_test = [], []
for i in range(0, len(df_test), 300):
# for key, df_group in df_test_group:
    group_val = df_test[i:i+300][fea_col].values
    X_test.append(group_val)
    Y_test.append(df_test.iloc[i]['label'])
```

```python
X_test = np.array(X_test)
X_test.shape
```

```python
from torch.utils.data import DataLoader,TensorDataset
import torch
X_test = torch.tensor(X_test, dtype=torch.float)
Y_test = torch.tensor(Y_test, dtype=torch.float)
testloader = DataLoader(TensorDataset(X_test, Y_test), 1, False)
```

```python
import torch
from torch import nn
# LSTM模型搭建
class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, output_size=387, num_layers=2):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers,batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)
    def forward(self, _x):
        x, _ = self.lstm(_x)
        x1 = self.fc(x[:,-1])
        return x1
```

```python
from torch import optim, nn
from tqdm import tqdm
# 模型构建
modelDNN = LSTM(4, 32, output_size=387, num_layers=2)
optimizer = optim.Adam(modelDNN.parameters())
criterion = nn.CrossEntropyLoss()
best_losss = 10
# 训练轮数
for i in range(30):
    pbar = tqdm(trainloader)
    for X, y in pbar:
        # 优化
        y_pred = modelDNN(X)
#         print(y_pred.size(), y.size())
        
        loss = criterion(y_pred, y.long())
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        pbar.set_description(f'loss: {loss.item()}')
    if loss.item() < best_losss:
        torch.save(modelDNN.state_dict(), 'model.th')
        best_losss = loss.item()
```

```python
from torch import optim, nn
from tqdm import tqdm
# 模型构建
modelDNN = LSTM(4, 32, output_size=387, num_layers=2)
optimizer = optim.Adam(modelDNN.parameters())
criterion = nn.CrossEntropyLoss()

submit = []
modelDNN.load_state_dict(torch.load('model.th'))
modelDNN = modelDNN.cuda()
pbar = tqdm(testloader)
with torch.no_grad():
    for qid, (X, y) in enumerate(pbar):
        X = X.cuda()
        y_pred = torch.softmax(modelDNN(X), dim=-1)
        # print(y)
        # print(y_pred[:,y.long()[0]].data.numpy()[0])
        submit.append([qid+1, y_pred[:,y.long()[0]].cpu().data.numpy()[0]])
```
