```python
def z_score(X):
    return (X - X.mean(axis=0))/X.std(axis=0)
```

```python
# material 3과 material 5를 새 feature로 선정
train["m3_missing"] = train["measurement_3"].isnull().astype(np.int8)
train["m5_missing"] = train["measurement_5"].isnull().astype(np.int8)

test["m3_missing"] = test["measurement_3"].isnull().astype(np.int8)
test["m5_missing"] = test["measurement_5"].isnull().astype(np.int8)
```

```python
train.iloc[:, 26:28]
```

```python
train["measurement_avg"] = train[[f"measurement_{i}" for i in range(3,17)]].mean(axis=1)
test["measurement_avg"] = test[[f"measurement_{i}" for i in range(3,17)]].mean(axis=1)
```

```python
# 결측값 대체
train.fillna(train.mean(), inplace=True)
test.fillna(test.mean(), inplace=True)
```

```python
# product code colum 삭제
train = train.drop(['product_code'], axis=1)
test = test.drop(['product_code'], axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
```

```python
attribute = ["attribute_0","attribute_1"]
enc = OneHotEncoder(handle_unknown='ignore')
enc.fit(train[attribute])
enc.fit(test[attribute])
```

```python
le = LabelEncoder()
train["attribute_0"] = le.fit_transform(train["attribute_0"])
train["attribute_1"] = le.fit_transform(train["attribute_1"])

test["attribute_0"] = le.fit_transform(test["attribute_0"])
test["attribute_1"] = le.fit_transform(test["attribute_1"])
```

```python
train.head()
```

```python
train['area'] = train['attribute_2'] * train['attribute_3']
test['area'] = test['attribute_2'] * test['attribute_3']

cols = [col for col in train.columns if col not in ["id","product_code",'area',"failure","loading","attribute_2","attribute_3","attribute_1","attribute_0"]]

attributes = ['area', 'attribute_2', 'attribute_3', 'measurement_4', 'measurement_5', 'measurement_6']

train[attributes] = preprocessing.normalize(train[attributes])
test[attributes] = preprocessing.normalize(test[attributes])
```

```python
train[cols] = z_score(train[cols])
test[cols] = z_score(test[cols])
```

```python
import os
import sys
import joblib
import numpy as np
import pandas as pd
import gc; gc.enable()
from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score
from sklearn.naive_bayes import GaussianNB
# from feature_engine.encoding import WoEEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from kuma_utils.preprocessing.imputer import LGBMImputer
import warnings; warnings.filterwarnings("ignore")
```

```python
params = {"max_iter": 500, "C": 0.0001, "penalty": "l2", "solver": "newton-cg"}

oof = np.zeros(len(train))
test_preds = np.zeros(len(test))
# Grooup KFold가 별로 좋은 성능이 아닝서 StartifieldKFold로 변경
for fold, (train_idx, val_idx) in enumerate(StratifiedKFold(n_splits = 5, shuffle = True, random_state = 0).split(train, train["failure"])):
    x_train, y_train = train.loc[train_idx][features], train.loc[train_idx]["failure"]
    x_val, y_val = train.loc[val_idx][features], train.loc[val_idx]["failure"]

    model = LogisticRegression(**params)
    model.fit(x_train, y_train)
    y_pred_1 = model.predict_proba(x_val)[:, 1]
    test_preds_1 = model.predict_proba(test[features])[:, 1] / 5
       
    lgb_params = {
        'seed': 42,
        'n_jobs': -1,
        'lambda_l2': 2,
        'metric': "auc",
        'max_depth': -1,
        'num_leaves': 100,
        'boosting': 'gbdt',
        'bagging_freq': 10,
        'learning_rate': 0.01,
        'objective': 'binary',
        'min_data_in_leaf': 40,
        'num_boost_round': 1000,
        'feature_fraction': 0.95,
        'bagging_fraction': 0.95,
    }
    
    model = LGBMClassifier(**lgb_params)
    model.fit(x_train, y_train, eval_set = [(x_val, y_val)], early_stopping_rounds = 30)            
    y_pred_2 = model.predict_proba(x_val)[:, 1]
    test_preds_2 = model.predict_proba(test[features])[:, 1] / 5
        
#     model = GaussianNB(var_smoothing=0.5, priors=[len(y_train[y_train == 0]) / len(y_train), len(y_train[y_train == 1])/len(y_train)])
#     model.fit(x_train, y_train)
#     y_pred_3 = model.predict_proba(x_val)[:, 1]
#     test_preds_3 = model.predict_proba(test_data[features])[:, 1] / 5
    
    oof[val_idx] = (y_pred_1 * 0.9)     + (0.1 * y_pred_2)
    test_preds   = (test_preds_1 * 0.9) + (0.1 * test_preds_2)
    
    print(f"Val score: {roc_auc_score(y_val, oof[val_idx]):.7f}")

print(f"Val score: {roc_auc_score(train['failure'], oof):.7f}")
```
