```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
#glob와 비슷한 자매품으로 pickle이 존재
# glob함수는 인자로 받은 패턴과 일치하는 모든 파일과 디렉터리의 리스트를 반환
# 패턴을 *로 주면 모든 파일과 디렉토리를 보기 가능
from tqdm import tqdm

import tensorflow as tf
from tensorflow.keras.models import load_model, Model
from tensorflow.keras.layers import Conv3D, ConvLSTM2D, BatchNormalization, Input
from tensorflow.keras.optimizers import Adam

#strategy = tf.distribute.MirroredStrategy()
strategy = tf.distribute.get_strategy()


# 데이터 가져오기
path= '/home/centos/suhyeok/seaice/input/'
train_csv = pd.read_csv(path+'train.csv')
print(train_csv.head())

train_path = sorted(glob(path+'train/*.npy'))
#list sort가 아닌 데이터를 정렬해주는 sorted
#print(len(train_path))
# 12달 30년 잘라주기
train_path = train_path[-30*12:]
#print(len(train_path))

#tqdm을 통해 train list에 train_path의 load한 값을 append한 것을 시각화
train = []
for p in tqdm(train_path):
    train.append(np.load(p))
# np.array를 통해서 ndarray로 바꿔줌
train = np.array(train)

print(train.shape)
# train의 shape는 (480, 448, 304, 5) 4차원
# 갯수, 높이, 너비

# arr_4 = np.arange(1,13).reshape(2,3,2,1)
# print(arr_4.shape)
# print(arr_4)
# 4차원 찍어본 결과
# shape (3,2,2,2) shape (2,3,2,2)   shape(2,2,3,2)      shape(2,2,2,3)
#[[[[ 1  2]     [[[[1 2]            [[[[1 2]            [[[1 2 3]
# [ 3  4]]          [3 4]]              [3 4]              [4 5 6]]
#                                       [5 6]]
# [[ 5  6]          [[5 6]                               [[7 8 9]
# [ 7  8]]]         [7 8]]              [[7 8]             [10 11 12]]]
#                                        [9 10]
# [[[ 9 10]         [[9 10]             [11 12]]]       [[[13 14 15]
# [11 12]]          [11 12]]]                              [16 17 18]]
#                                   [[[13 14]
# [[13 14]          [[[13 14]          [15 16]          [[19 20 21]
# [15 16]]]           [15 16]]          [17 18]]        [22 23 24]]]]
#
# [[[17 18]         [[17 18]        [[19 20]
# [19 20]]          [19 20]]        [21 22]
#                                   [23 24]]]]
# [[21 22]          [[21 22]
# [23 24]]]]        [23 24]]]]

plt.figure(figsize=(15,10))
# plt.figure는 새로운 figure을 생성해줌
# figsize는 가로길이 15인치 세로길이 10인치
plt.subplot(151)
# subplot은 하나의 figure에 표처럼 여러 axes를 표현 가능하게 해줌
plt.imshow(train[0,:,:,0])
plt.subplot(152)
plt.imshow(train[0,:,:,1])
plt.subplot(153)
plt.imshow(train[0,:,:,2])
plt.subplot(154)
plt.imshow(train[0,:,:,3])
plt.subplot(155)
plt.imshow(train[0,:,:,4])
plt.show()


# 0은 해빙 농도, 1은 북극점(위성 관측 불가 지점),2는 해안선마스크,3은 육지 마스크,4는 결측값 총 5개의 채널로 구성
# 각 채널은 448*304로 구성
# 1에서 북극점은 위성이 관측 불가한 지점이어서 구멍이 뚫려있음
# 예측시 채널1+채널2하면 됨  


# 월 별로 데이터 분리
# slicing a[start:end:step]

train_month = []
for i in range(12):
    train_month.append(train[i::12, :,:])

train_month = np.array(train_month)

print(train_month.shape)
# (12,40,448,304,5)

# 이미지를 잘랐을 때 사이즈가 같도록 padding
stride = 32
temp = np.zeros([train_month.shape[0],
                 train_month.shape[1],
                 train_month.shape[2]+(stride-train_month.shape[2]%stride),
                 train_month.shape[3]+(stride-train_month.shape[3]%stride),
                 train_month.shape[4]], np.uint8)
# 크기가 늘어난 train_month.shape[2]와 train_month.shape[3]에 원래 값을 넣어주고 이후로는 0값

#wide_pad = int((stride-train_month[2]%stride)/2)
#height_pad = int((stride-train_month[3]%stride/3))
#temp[:,:,wide_pad:wide_pad+train_month[2],height_pad:height_pad+train_month.shape[3],:]=train_month
temp[:, :,:train_month.shape[2], :train_month.shape[3], :] = train_month
train_month = temp
del temp

# window_size는 과거 5년을 기반으로 예측하겠다는 것을 의미
window_size = 5

x_train = []
y_train = []
#train_month.shape[1]-window_size : 35
for i in range(train_month.shape[1]-window_size):
    # x_train에 i부터 i+window_size까지 5개 들어감
    x_train.append(train_month[:, i:i+window_size, :,:,:])
    # y_train에는 i+window_size부터 i+window_size+1까지 1개 들어감
    y_train.append(train_month[:, i+window_size:i+window_size+1, :,:,:])

x_train = np.array(x_train)
y_train = np.array(y_train)

print("x_train.shape, y_train.shape")
print(x_train.shape, y_train.shape)


#특정 달의 5년간 변화와 그 다음 해의 상태
m = 1
plt.figure(figsize=(10,15))
for y in range(x_train.shape[2]):
    plt.subplot(1,window_size,y+1)
    plt.imshow(x_train[0,m-1,y,:,:,0])
    plt.show()

plt.imshow(y_train[0,m-1,-1,:,:,0])
plt.show()

# 이미지가 크면 ConvLSTM 모델의 메모리가 부족하는 현상 발생
# 가로세로 32px씩 이동하며 128X128로 이미지 자름
feature_size =128
x_train_ = []
y_train_ = []
for y in range(x_train.shape[0]):
    for m in range(12):
        for i in range((x_train.shape[3]-feature_size)//stride+1): #14
            for j in range((x_train.shape[4]-feature_size)//stride+1): #9
                x_ = x_train[y, m, :, stride*i:stride*i+feature_size, stride*j:stride*j+feature_size, :1]
                # 왜 (35,12,5,480,320,5)이던 x_train이 x_의 shape가 (5,64,64,1)인거지
                y_ = y_train[y, m, :, stride*i:stride*i+feature_size, stride*j:stride*j+feature_size, :1]
                x_train_.append(x_)
                y_train_.append(y_)
#for 문 35 * 12 * 14 * 9 다 돌면 52920
x_train_ = np.array(x_train_)
print(x_train_.shape)
y_train_ = np.array(y_train_)
#
x_val_ = []
y_val_ = []

for m in range(12):
    for i in range((x_train.shape[3]-feature_size)//stride+1):
        for j in range((x_train.shape[4]-feature_size)//stride+1):
            x_ = x_train[-1, m, :, stride*i:stride*i+feature_size, stride*j:stride*j+feature_size, :1]
            y_ = y_train[-1, m, :, stride*i:stride*i+feature_size, stride*j:stride*j+feature_size, :1]
            x_val_.append(x_)
            y_val_.append(y_)

x_val_ = np.array(x_val_)
y_val_ = np.array(y_val_)
#
print(x_train_.shape, y_train_.shape, x_val_.shape, y_val_.shape)
#
i = 22
plt.figure(figsize=(15,10))
for y in range(x_train_.shape[1]):
    plt.subplot(1,window_size,y+1)
    plt.imshow(x_train_[i,y,:,:,0])
plt.show()
plt.imshow(y_train_[i,-1,:,:,0])

def train_map_func(x, y):
    x = tf.cast(x, tf.float32)/250
    y = tf.cast(y, tf.float32)
    return x, y


# Dataset 구축
BATCH_SIZE = 16
# x_train_, y_train_을 tensorflow로 넣는다
train_dataset = tf.data.Dataset.from_tensor_slices((x_train_, y_train_))
train_dataset = train_dataset.map(train_map_func, num_parallel_calls=tf.data.experimental.AUTOTUNE)
train_dataset = train_dataset.batch(BATCH_SIZE)
train_dataset = train_dataset.prefetch(buffer_size=tf.data.experimental.AUTOTUNE)

val_dataset = tf.data.Dataset.from_tensor_slices((x_val_, y_val_))
val_dataset = val_dataset.map(train_map_func, num_parallel_calls=tf.data.experimental.AUTOTUNE)
val_dataset = val_dataset.batch(BATCH_SIZE)
val_dataset = val_dataset.prefetch(buffer_size=tf.data.experimental.AUTOTUNE)

print(next(iter(train_dataset))[0].shape, next(iter(train_dataset))[1].shape)
print(next(iter(val_dataset))[0].shape, next(iter(val_dataset))[1].shape)

# Modeling
with strategy.scope():
    inputs=Input((None, x_train_.shape[2], x_train_.shape[3], 1))

    convLSTM = ConvLSTM2D(filters=64, kernel_size=3, padding='same', return_sequences=True)(inputs)
    convLSTM = BatchNormalization()(convLSTM)

    convLSTM = ConvLSTM2D(filters=64, kernel_size=3, padding='same', return_sequences=True)(convLSTM)
    convLSTM = BatchNormalization()(convLSTM)

    convLSTM = ConvLSTM2D(filters=64, kernel_size=3, padding='same', return_sequences=True)(convLSTM)
    convLSTM = BatchNormalization()(convLSTM)

    convLSTM = ConvLSTM2D(filters=32, kernel_size=3, padding='same', return_sequences=True)(convLSTM)
    convLSTM = BatchNormalization()(convLSTM)

    convLSTM = ConvLSTM2D(filters=32, kernel_size=3, padding='same', return_sequences=True)(convLSTM)
    convLSTM = BatchNormalization()(convLSTM)

    outputs = Conv3D(filters=1, kernel_size=3, activation='relu', padding='same', data_format='channels_last')(convLSTM)

    model = Model(inputs, outputs)
    model.compile(loss='mae', optimizer=Adam())

model.summary()

if not (os.path.isdir('models')):
    os.makedirs('models')

callbacks_list = [
    tf.keras.callbacks.ModelCheckpoint(
        filepath = './models/convlstm_model.h5',
        monitor='val_loss',
        save_best_only=True
    )
]
#
hist = model.fit(train_dataset, epochs=150, validation_data=val_dataset, callbacks=callbacks_list)

plt.plot(hist.history['loss'][1:])
plt.plot(hist.history['val_loss'][1:])
plt.show()

model = load_model('./models/convlstm_model.h5')

pred = model.predict(next(iter(val_dataset))[0])
pred = np.where(pred>250, 250, pred)

for i in range(BATCH_SIZE):
    plt.subplot(141)
    plt.imshow(next(iter(val_dataset))[1][i,0,:,:,0])
    plt.subplot(142)
    plt.imshow(pred[i,0,:,:,0])
    plt.show()
    print(np.mean(np.abs(next(iter(val_dataset))[1][i,0,:,:,0] - pred[i,0,:,:,0])))

    

# 2014~2018년 5년간의 데이터를 학습과 같은 형태로 전처리
x_test = train_month[:,-1*window_size:,:,:,:1]
x_test = np.swapaxes(x_test , 0, 1)

x_test_ = []
for m in range(12):
    feature = []
    for i in range((x_train.shape[3]-feature_size)//stride+1):
        for j in range((x_train.shape[4]-feature_size)//stride+1):
            x_ = x_test[:, m, stride*i:stride*i+feature_size, stride*j:stride*j+feature_size, :1]/250
            feature.append(x_)
    feature = np.array(feature)
    x_test_.append(feature)
x_test_ = np.array(x_test_)

# 2015~2019년 데이터 또한 학습과 같은 형태로 전처리
x_test_20 = train_month[:,-1*window_size:,:,:,:1]
x_test_20 = np.swapaxes(x_test_20 , 0, 1)
x_test_20_ = []
for m in range(12):
    feature = []
    for i in range((x_train.shape[3]-feature_size)//stride+1):
        for j in range((x_train.shape[4]-feature_size)//stride+1):
            x_ = x_test_20[:, m, stride*i:stride*i+feature_size, stride*j:stride*j+feature_size, :1]/250
            feature.append(x_)
    feature = np.array(feature)
    x_test_20_.append(feature)
x_test_20_ = np.array(x_test_20_)

print(x_test_.shape,x_test_20.shape)

# 추론 후 월별로 저장
preds = []
for m in tqdm(range(12)):
    pred = model.predict(x_test_[m])
    pred = np.where(pred>250, 250, pred)
    preds.append(pred)


# 추론 결과 중복된 영역 평균 이용해 앙상블
voting_mask = np.zeros((x_train.shape[-3], x_train.shape[-2]))
ones = np.ones((feature_size, feature_size))

for i in range((x_train.shape[3]-feature_size)//stride+1):
    for j in range((x_train.shape[4]-feature_size)//stride+1):
        voting_mask[stride*i:stride*i+feature_size, stride*j:stride*j+feature_size] += ones

plt.imshow(voting_mask)
plt.show()

results = []
for m in range(12):
    pred = np.zeros((x_train.shape[-3], x_train.shape[-2]))
    k = 0
    for i in range((x_train.shape[3]-feature_size)//stride+1):
        for j in range((x_train.shape[4]-feature_size)//stride+1):
            pred[stride*i:stride*i+feature_size, stride*j:stride*j+feature_size] += preds[m][k,-1,:,:,0]
            k+=1
    pred /= voting_mask
    results.append(pred[:448, :304])
results = np.array(results)

# 추론 결과 plot
for m in range(12):
    plt.imshow(results[m])
    plt.show()
    print(m+1, '\n')

submission = pd.read_csv(path+'sample_submission.csv')

sub_2019 = submission.loc[:11, ['month']].copy()
sub_2020 = submission.loc[12:, ['month']].copy()
#
sub_2019 = pd.concat([sub_2019, (pd.DataFrame(results.reshape([12,-1])))], axis=1)
sub_2019.columns = submission.columns
sub_2020 = pd.concat([sub_2020, (pd.DataFrame(results.reshape([12,-1]), index=sub_2020.index))], axis=1)
sub_2020.columns = submission.columns
submission = pd.concat([sub_2019, sub_2020])
#
submission.to_csv('submission_3.csv', index=False)
```
